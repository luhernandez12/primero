```python
#lista de todos los animales en la veterinaria
lista_animales=[]
cont_codig_ident = 1


#informacion basica del animal y asignacion de codigo por especie
while True:
  print(f"Bienvenido a la veterinaria de la UdeA".center(90))

  print("Escoja la opcion:")
  print("1. Ingresar infomacion del animal ")
  print("2. Lista de pacientes ")
  print("3. Cantidad de pacientes ")
  print("4. Promedio de edades de los animales ")
  print("5. Estado de los animales ")
  print("6. Buscar un paciente por su codigo de identificacion ")
  print("7. Salir")
  menu=int(input("Ingresar la opcion a la cual quiere acceder "))

   # PRIMERA OPCION
  if menu==1:
    animales=[]
    # Informacion basica del animal
    nombre_animal=input("Ingresar el nombre del animal:")
    animales.append(nombre_animal)
    edad=int(input("ingresar la edad de su animal: "))
    animales.append(edad)
    especie=int(input("Tipo de animal\n1.Gato\n2.Perro\nIntroducir el numero correspondite al tipo de animal que tiene: "))
    if especie==1:
      tipo_animal='Gato'
    else:
      tipo_animal='Perro'
    animales.append(tipo_animal)




    # estableciendo codigo del animal

    if especie == 1:
        codigo_identificacion = "CAN0{}".format(cont_codig_ident)
    elif especie == 2:
        codigo_identificacion = "FEL0{}".format(cont_codig_ident)
    else:
        print("Entrada no válida.")
        continue

    animales.append(codigo_identificacion)
    cont_codig_ident += 1

    # Estableciendo condicon del animal
    condicon_animal=int(input("Estado del animal:\n1.Grave\n2.Critico\nIngresar el numero que indique el estado de su animal "))
    if condicon_animal==1:
      estado='Grave'
    else:
      estado='Critico'
    animales.append(estado)
    print(estado)
    lista_animales.append(animales)
    print("codigo_identificacion",codigo_identificacion)
    print("\n")


  #SEGUNDA OPCION
  elif menu==2:
    # Informacin lista de pacientes
   print("\n")
   print("LISTA DE INFOMACION DE ANIMALES:")
   for animales in lista_animales:
      print(animales)
   print("--------------")


  #TERCERA OPCION
  elif menu==3:
    #Cantidad de perros y gatos
    print("\n")
    print("CANTIDAD DE ANIMALES INGRESADOS SEGUN SU ESPECIE")

    #Gatos
    gato='Gato'
    contador = 0
    for animales in lista_animales:
        for genero in animales:
            if genero == gato:
                contador += 1
    print(f"Cantidad de gatos ingresados:{contador}")

    #Perros
    perro='Perro'
    contador_1 = 0

    for animales in lista_animales:
        for genero_1 in animales:
            if genero_1 == perro:
                contador_1 += 1
    print(f"Cantidad de perros ingresados:{contador_1}")
    print("--------------")
    print("\n")


 #CUARTA OPCION
 # promedio de edades animales
  elif menu==4:
    suma = 0
    for animales in lista_animales:
        suma += animales[1]

    promedio = suma / len(lista_animales)
    print("El promedio de las edades de los animales es de:", promedio)
    print("--------------")
    print("\n")

 #QUINTA OPCION
 #Cantidad de animales en esatdo grave y critico
  elif menu==5:
    print("\n")
    print("CANTIDAD DE ANIMALES INGRESADOS SEGUN SU ESTADO")

    #Graves
    graves='Grave'
    cont_1 = 0
    for animales in lista_animales:
        for s in animales:
            if s == graves:
                cont_1 += 1
    print(f"Cantidad animales ingresados en estado grave:{cont_1}")

    #Critricos
    critico='Critico'
    cont_2 = 0

    for animales in lista_animales:
        for l in animales:
            if l == critico:
                cont_2 += 1
    print(f"Cantidad animales ingresados en estado critico:{cont_2}")
    print("--------------")

 # SEXTA OPCION
  elif menu==6:
    buscar=input("igrese el codigo  de idetificacion del animal(debe comenzar con CAN O FEL)  ")

    if not (buscar.startswith("CAN") or buscar.startswith("FEL")):
        print("Formato de código inválido.")
        continue

    encontrado = False
    for animal in lista_animales:
      if buscar in animal:
          print("Animal correspondiente a la identificación ingresada:")
          indice_codigo = animal.index(buscar)
          print(f"Nombre: {animal[indice_codigo - 3]}\nEdad: {animal[indice_codigo - 2]}\nEspecie: {animal[indice_codigo - 1]}\nCódigo de identificación: {animal[indice_codigo]}\nEstado: {animal[indice_codigo + 1]}")
          encontrado = True
          break

    if not encontrado:
        print(f"El código {buscar} no está en la base de datos. Inténtelo otra vez.")


 #septima opcion
  else:
    fallos=0
    while fallos<3:
      submenu=input("\n1.Aceptar\n.2.Rechazar\n")
      if submenu=="1" or submenu=="2":
         break
      else:
        fallos+=1
        print("Ingresar la opcion correcta")
    if submenu=="1":
       break
    else:
      continue








```
